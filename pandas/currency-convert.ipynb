```python
import pandas as pd
import numpy as np
import requests as req
```

```python
with open("./src/keys.txt") as keys:
    api_keys = {}
    for line in keys.readlines():
        api_keys[line.split(",")[0]]=line.split(",")[1].strip()
```

```python
data_size = 5000
transactions = pd.DataFrame({"CNIC": np.random.randint(1,1000, size=(data_size)),
 "TransactionID":np.random.randint(1,2, size=(data_size)),
 "Transaction Date":pd.date_range('1690-01-01', periods=data_size, freq="M"),
 "Description":np.random.choice(["TopUP","IBFT","FT"], size=(data_size)),
 "City":np.random.choice(["Karachi","Islamabad","Lahore"], size=(data_size)),
 "Amount": np.random.randint(1,1000000, size=(data_size)),
 "Number":np.random.choice(['036',"051","104",'840'], size=(data_size))})
transactions
```

```python
currInfo = pd.read_html("https://www.iban.com/currency-codes")[0].dropna().reset_index(drop=True)
currInfo
```

```python
currInfo.Number = currInfo.Number.astype("int")
currInfo = currInfo.drop_duplicates(subset=["Number"]).reset_index(drop=True)
currInfo
```

```python
transactions.Number = transactions.Number.astype("int")
transactions
```

```python
u_trans_curr_no = transactions.Number.unique()
u_trans_curr_no
```

```python
df_trans_curr_prog = transactions.merge(currInfo,on="Number")
df_trans_curr_prog
```

```python
currency_world = pd.read_json("https://api.currencyfreaks.com/latest?apikey={0}".format(api_keys["currencyfreaks"]))
currency_world.loc["AMD"]
```

```python
convrt_str = "USD_PKR"
res = req.get("https://free.currconv.com/api/v7/convert?q={0}&compact=ultra&apiKey={1}".format(convrt_str,api_keys["currconv"]))
```

```python
PKR_USD = res.json()[convrt_str]
PKR_USD
```

```python
curr = currency_world.drop(["date","base"],axis=1)
curr.loc["AMD"]
```

```python
curr["Rates_In_USD"] = 1.0 / curr["rates"] 
curr["USD_Rate_In_PK"] = curr["Rates_In_USD"] * PKR_USD
```

```python
curr.loc["AMD"]
```

```python
df_trans_curr = df_trans_curr_prog.merge(curr,left_on="Code",right_index=True)
df_trans_curr
```

```python
df_trans_curr["Amount_PKR"] = df_trans_curr.Amount * df_trans_curr.USD_Rate_In_PK
df_trans_curr
```
