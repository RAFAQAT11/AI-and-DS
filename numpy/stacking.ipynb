```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
img = Image.open("./assets/test1.jpg")
img_arr = np.array(img)
print(img.size,img_arr.shape)
```

```python
plt.imshow(img_arr)
plt.show()
```

```python
img2 = Image.open("./assets/test2.jpg")
img_arr2 = np.array(img2)
print(img2.size,img_arr2.shape)
```

```python
plt.imshow(img_arr2)
plt.show()
```

```python
def crop_rgb(img,crop_x,crop_y,center=True):
    """
        img: numpy array
        crop_x: number of pixels for x dim
        crop_y: number of pixels for y dim
        center: true for center
    """
    y, x, _ = img.shape
    start_y, end_y, start_x, end_x = \
    (y//2 - crop_y, y//2 + crop_y, x//2 - crop_x, x//2 + crop_x) \
    if center else \
    (0, crop_y, 0, crop_x)
    
    return img[start_y:end_y,start_x:end_x,:]
```

```python
croped1 = crop_rgb(img_arr,1000,500,False)
croped2 = crop_rgb(img_arr2,1000,500,False)

print(croped1.shape)
print(croped2.shape)
```

```python
plt.imshow(croped1)
plt.show()
```

```python
plt.imshow(croped2)
plt.show()
```

```python
hStacked = np.hstack((croped1,croped2))
```

```python
plt.imshow(hStacked)
plt.show()
```

```python
vStacked = np.vstack((hStacked,hStacked))
```

```python
plt.imshow(vStacked)
plt.show()
```
