```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
img = Image.open("./assets/fb.jpg")
img_arr = np.array(img)
print(img.size,img_arr.shape)
```

```python
plt.imshow(img_arr)
plt.show()
```

```python
def crop_rgb(img,crop_x,crop_y,center=True):
    """
        img: numpy array
        crop_x: number of pixels for x dim
        crop_y: number of pixels for y dim
        center: true for center
    """
    y, x, _ = img.shape
    start_y, end_y, start_x, end_x = \
    (y//2 - crop_y, y//2 + crop_y, x//2 - crop_x, x//2 + crop_x) \
    if center else \
    (0, crop_y, 0, crop_x)
    
    return img[start_y:end_y,start_x:end_x,:]
```

```python
croped = crop_rgb(img_arr,250,250)
croped.shape
```

```python
plt.imshow(croped)
plt.show()
```
