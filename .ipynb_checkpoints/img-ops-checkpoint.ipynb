```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
img = Image.open("./numpy-04321/assets/fb.jpg")
oimg_arr = np.array(img)
print(img.size,oimg_arr.shape)
```

```python
plt.imshow(oimg_arr)
plt.show()
```

```python
def div(img,scaler):
    return img/scaler

def mul(img,scaler):
    return img * scaler

def add(img,scaler):
    return img+scaler

def minus(img,scaler):
    return img - scaler

def display(img):
    plt.imshow(img)
    plt.show()
```

```python
normImg = div(oimg_arr,255)

display(div(normImg,normImg.std()))
```

```python
display(div(normImg,normImg.mean()))
```
