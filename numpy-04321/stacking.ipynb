```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
img = Image.open("./assets/test1.jpg")
img_arr = np.array(img)
print(img.size,img_arr.shape)
```

```python
plt.imshow(img_arr)
plt.show()
```

```python
img2 = Image.open("./assets/test2.jpg")
img_arr2 = np.array(img2)
print(img2.size,img_arr2.shape)
```

```python
plt.imshow(img_arr2)
plt.show()
```

```python
croped1 = img_arr[0:500,0:1000,:]
croped2 = img_arr2[0:500,0:1000,:]
```

```python
print(croped1.shape)
print(croped2.shape)
```

```python
plt.imshow(croped1)
plt.show()
```

```python
plt.imshow(croped2)
plt.show()
```

```python
hStacked = np.hstack((croped1,croped2))
```

```python
plt.imshow(hStacked)
plt.show()
```

```python
vStacked = np.vstack((hStacked,hStacked))
```

```python
plt.imshow(vStacked)
plt.show()
```
