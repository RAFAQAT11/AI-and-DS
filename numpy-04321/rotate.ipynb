```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def display(img):
    plt.imshow(img,cmap="gray")
    plt.show()
    
def convertToGray(img):
    return (0.212*img[:,:,0]+0.701*img[:,:,1]+0.087*img[:,:,2])/225.0
```

```python
img = plt.imread("./assets/cheetah.jpg")
img_gray = convertToGray(img)
display(img_gray[:,::-1])
```
