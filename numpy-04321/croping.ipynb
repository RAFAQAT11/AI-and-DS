```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
img = Image.open("./assets/test1.jpg")
img_arr = np.array(img)
print(img.size,img_arr.shape)
```

```python
plt.imshow(img_arr)
plt.show()
```

```python
v_center = np.floor(img_arr.shape[0]*0.5)
h_center = np.floor(img_arr.shape[1]*0.5)
print(v_center,h_center)
crop_start_x = int( v_center - 250)
crop_end_x = int( v_center + 250)

crop_start_y = int( h_center - 250)
crop_end_y = int( h_center + 250)
```

```python
croped = img_arr[crop_start_x:crop_end_x,crop_start_y:crop_end_y,:]
croped.shape
```

```python
plt.imshow(croped)
plt.show()
```
