```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from numpy.random import randint
import os
import imageio
import pandas as pd
```

```python
def crop_img(img,x,y,w,h):
    return img[y:y+h,x:x+w]
```

```python
imags_path = []
templates_path = []
Top_Left_X = []
Top_Left_Y = []
Bottom_Right_X = []
Bottom_Right_Y = []

WIDTH=100
HEIGHT=100

record = []
imgs = os.listdir("./downloads/images/")

np.random.seed(100)

for img in imgs:
    
    img_org = cv2.imread(f"./downloads/images/{img}")
    img_main = cv2.cvtColor(img_org, cv2.COLOR_BGR2GRAY)
    
    while True:
        var = randint(low = -200, high = 200, size = 1)
        if var[0] not in record:
            break
    
#     x,y,w,h = 200+var[0],200+var[0],100+var[0],100+var[0]
    
    x,y,w,h = 200+var[0],200+var[0],WIDTH,HEIGHT
    
    temp=crop_img(img_org,x,y,w,h)
#     plt.imshow(temp,cmap="gray"),plt.show()

    
    imageio.imwrite(f"./downloads/templates/template-{img}",cv2.cvtColor(temp, cv2.COLOR_BGR2RGB))
        
    imags_path.append(f"./downloads/images/{img}")
    templates_path.append(f"./downloads/templates/template-{img}")
    Top_Left_X.append(x)
    Top_Left_Y.append(y)
    Bottom_Right_X.append(w+x)
    Bottom_Right_Y.append(h+y)
    
    record.append(var[0])

positive=["Positive"]*50
negative=["Negative"]*50
actual_values = positive+negative
wrong_templates = templates_path[::-1]


dataSet = {
    "Images":imags_path+imags_path,
    "Templates":templates_path+wrong_templates,
    "Top_Left_X":Top_Left_X+Top_Left_X,
    "Top_Left_Y":Top_Left_Y+Top_Left_Y,
    "Bottom_Right_X":Bottom_Right_X+Bottom_Right_X,
    "Bottom_Right_Y":Bottom_Right_Y+Bottom_Right_Y,
    "Actual_values":actual_values
}
df = pd.DataFrame(dataSet)
```

```python
df.to_csv("dataSet.csv",index=False)
```
